###### Determining colinearity

Colinearity is the presence of a linear relationsihp between predictor variables in a regression model and undermines the power the regression. In relation to the admissions dataset we can visualy inspect a scatterplot matrix to determine if there are linear relatiionships between predictors

```python
pd.tools.plotting.scatter_matrix(df,alpha=0.2,figsize=(6,6),diagonal='hist')

```

##### The above indicates that there may be some positive linear correlation with GPA and GRE meaning one of the variables would have to be eliminated from any further analysis
